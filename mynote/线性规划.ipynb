### 线性规划
$$
目标函数：z=min(c^Tx) \newline
约束条件=
\begin{cases}
A_{ub}x\leq b_{ub}\\
A_{eq}*x=b{eq}\\
l \leq x \leq u
\end{cases}
$$
#### 整数规划
线性规划问题的最优解可能是分数或小数。整数规划是指变量的取值只能是整数的规划
#### 0-1规划
0-1 整数规划是一类特殊的整数规划，变量的取值只能是 0 或 1。主要用于求解互斥的决策问题、互斥的约束条件问题、固定费用问题和分派问题  
* 决策变量互斥  
例：背包问题
$$
maxz=\sum_{i=1}^{i=n} v_i*x_i \newline
\begin{cases}
\sum w_ix_i\leq B\\
x_i=\begin{cases}
0 &不选择\\
1 &选择\\
   \end{cases}
\end{cases}
$$  
例：指派问题  
n个人n个工作（每个人一个，每个工作一个人），第 i 个人做第 j 件事用时$c_{ij}$，使总用时最小  
$$
minz=\sum_{i=1}^n\sum_{j=1}^n(c_{ij}x_{ij}) \newline
\begin{cases}
\sum b_ix_i\leq B\\
x_{ij}=\begin{cases}
       0 &不选择\\
       1 &选择\\
    \end{cases}\\
\sum_{i=1}^n x_{ij}=1\\
\sum_{j=1}^n x_{ij}=1\\
\end{cases}
$$  
例：选址问题  
    * 中位问题  
    有 N 个候选服务站和 M 个需求点，要从 N 个候选服务站中选择 P 个，使所有需求点到最近的服务站的加权距离 $w_i*d_{ij}$ 和最小  
    $$
     minz=\sum_{i=1}^M \sum_{j=1}^N w_i*d_{ij}*y_{ij} \newline
     \begin{cases}
       x_{j}=\begin{cases}
           0 &服务站j未选中\\
           1 &服务站j被选中\\
           \end{cases}\\
       y_{ij}=\begin{cases}
           0 &需求点i由服务站j服务\\
           1 &需求点i不由服务站j服务\\
           \end{cases}\\
    \sum_{j=1}^N x_{j}=P\\
    \sum_{j=1}^N y_{ij}=1\\
    y_{ij}-x_j \leq 0
    \end{cases}
    $$
    * 中心问题  
    有 N 个候选服务站和 M 个需求点，要从 N个候选服务站中选择 P个，使任一需求点到最近的服务站的最大距离最小  
    $$
    minD
    同上，多一约束：\sum_{j=1}^N w_i*d_{ij}*y_{ij} \leq D
    $$
    * 集合覆盖问题  
    有 N 个候选服务站和 M 个需求点，已知每个服务站的服务范围（或服务容量），要从 N个候选服务站中选择若干个，使所有需求点得到服务（到所属服务站的距离或时间小于给定的临界值），服务站的个数最少或成本最小  
    $$
    服务站成本v_i\space\space 覆盖需求点i的服务站集合N_j=\{j:a_{ij}=1\}\newline
    minz=\sum_{j=1}^Nv_i*x_j\newline
    \begin{cases}
       x_{j}=\begin{cases}
           0 &服务站j未选中\\
           1 &服务站j被选中\\
           \end{cases}\\
       a_{ij}=\begin{cases}
           0 &需求点i被服务站j覆盖\\
           1 &需求点i不被服务站j覆盖\\
           \end{cases}\\
       \sum_{j\in N_j} x_{j}\geq1\\
    \end{cases}
    $$
    * 最大覆盖问题  
    已知服务站的数目和服务半径，设立 P个服务站使得可接受的服务需求最大  
    $$
    服务需求w_i\space\space 覆盖需求点i的服务站集合N_j=\{j:a_{ij}=1\}\newline
    \begin{gathered}
        \mathrm{max~\sum_{i\in N_i}w_iz_i}\\
        \mathrm{s.t.:~\begin{cases}
               z_i\leq\sum_{j\in N_i}x_j\\
               \sum_{j=1}^Nx_j=P\\
               x_j=\begin{cases}
                   0 &服务站j未选中\\
                   1 &服务站j被选中\\
                  \end{cases}\\
               z_i=\begin{cases}
                   0 &需求点i未被覆盖\\
                   1 &需求点i被覆盖\\
                   \end{cases}\\&\end{cases}}
    \end{gathered}$$
* 约束互斥  
多个互斥约束条件，只有x个起效  
m个约束条件  
$$
\sum_{i=1}^n a_{1,i}x_i \leq B_1 \newline
\sum_{i=1}^n a_{2,i}x_i \leq B_2 \newline
\sum_{i=1}^n a_{3,i}x_i \leq B_3 \newline
\sum_{i=1}^n a_{m,i}x_i \leq B_m
$$
构造后  
$$
\begin{cases}
\sum_{i=1}^n a_{1,i}x_i \leq B_1+(1-y_1)M \newline
\sum_{i=1}^n a_{2,i}x_i \leq B_2+(1-y_2)M \newline
\sum_{i=1}^n a_{3,i}x_i \leq B_3+(1-y_3)M \newline
\sum_{i=1}^n a_{m,i}x_i \leq B_n+(1-y_m)M \newline
y_1+y_2+y_3+···+y_m=x \newline
y_i=\begin{cases}
   0 & y_i不起效\\
   1 & y_i起效\\
   \end{cases}
\end{cases}
$$  
* 目标函数互斥  
例：固定费用问题  
$$
m种生产方式可供选择，采用第 j 种方式时的固定成本为 K_j，变动成本为 c_j，产量为x_j \newline 
对于第j钟方式：minz=K_j+c_j*x_j
$$  
构造后  
$$
minz=\sum_{j=1}^m(k_j*y_j+c_j*x_j) \newline
y_j=\begin{cases}
   0 &不选择\\
   1 &选择\\
   \end{cases}
$$

###### scipy库
$$
scipy.optimize.linprog(c, A_{ub}=None, b_{ub}=None, A_{eq}=None, b_{eq}=None, bounds=None, method='highs', callback=None, options=None, x0=None, integrality=None)
$$

例：
$$
maxz=2x_1+3x_2−5x_3 \newline 
\begin{cases}
-2x_1+5x_2-x_3\leq-10\\
x_1+3x_2+x_3\leq12\\
x_1+x_2+x_3=7\\
x_1,x_2,x_3\geq0
\end{cases}
$$

```python
import numpy as np
import scipy.optimize as op
z = np.array([2, 3, -5])

A_ub = np.array([[-2, 5, -1], [1, 3, 1]])
b_ub = np.array([-10, 12])
A_eq = np.array([[1, 1, 1]])
b_eq = np.array([7])
# 元组分别表示x1、x2和x3的边界范围
bounds = [0, None], [0, None], [0, None]
# 求目标函数取反后的最小值即目标函数的最大值
res = op.linprog(-z, A_ub, b_ub, A_eq, b_eq, bounds=bounds)
print(f"目标函数的最大值z={-res.fun}，此时目标函数的决策变量为{res.x}")
```

###### pulp库
* 通过pulp.LpProblem建立优化问题
$$
prob = pulp.LpProblem(name='NoName', sense=LpMinimize | LpMaximize)
$$
* 通过LpVariable函数创建优化变量
$$
pulp.LpVariable(name, lowBound=None, upBound=None, cat='Continuous', e=None)
$$
    * name：变量名，会输出到.lp文件
    * lowBound：变量下界
    * upBound：变量上界
    * cat：变量类型，可以为**LpInteger（整数规划）\ LpBinary（0-1规划）\ LpContinuous（线性规划）**三者之一（整型、二进制或连续）
    * e：指明变量是否在目标函数和约束中存在，主要用来实现列生成算法
* 通过+=向优化问题添加目标函数和约束
$$
prob += xxxxx,'目标函数'|'约束'
$$
* 使用求解器求解，默认为CBC
$$
prob.solve()
$$

```python
from pulp import *

prob = LpProblem('max_z', sense=LpMaximize)
x1 = LpVariable('x1', 0, None, LpInteger)
x2 = LpVariable('x2', 0, None, LpInteger)
x3 = LpVariable('x3', 0, None, LpInteger)

# 设置目标函数
prob += 2*x1+3*x2-5*x3
# 约束条件
prob += x1+x2+x3 == 7
prob += 2*x1-5*x2+x3 >= 10
prob += x1+3*x2+x3 <= 12

status = prob.solve()
# print("求解状态:", LpStatus[prob.status])
print(f"目标函数的最大值z={value(prob.objective)}，此时目标函数的决策变量为:",
      {v.name: v.varValue for v in prob.variables()})
```
