### 网络最大流问题（MFP）  
$$
V_i\underset{(C_{ij},x_{ij})}{\operatorname*{\longrightarrow}}V_j\\
V_i,V_j——节点\\
C_{ij}——从点i到点j的最大流量\\
x_{ij}——从点i到点j的当前流量\\
发点——仅有出弧\\
收点——仅有入弧\\
中间点——除收发点外的其他点
$$

##### 可行流  
* $0 \leq x_{ij}\leq C_{ij}$  
* 平衡：
    * 发点流量总输入=收点流量总输出
    * 中间点流量输入=输出

##### 增广链
可使流增大的链（表明当前链还有优化空间），表示如下：  
  
求解：  
$$\text{对于每一条弧}\begin{cases}\text{前向弧:找那些还可以增加流量的弧}\\\text{反向弧:找那些可以减少流量的弧}&\end{cases}$$

##### 最大流解法（Ford Fulkerson）  
$$\begin{aligned}
&\text{给发点}V_s\text{标号}(0,+\infty) \\
&\text{取一个已标号的点}V_i,\text{对于}V_i\text{一切未标号的邻接点}V_j：\\
&\phantom{xxxx} \text{如果有弧}(V_i,V_j)\in A\text{ 且}0\leqslant x_{ij}\leqslant C_{ij},\text{那么给}V_j标号(+V_i,\Delta_i)，\Delta_i=min\{C_{ij}-x_{ij},\Delta_i\} \\
&\phantom{xxxx} \text{如果有弧}(V_j,V_i)\in A\text{ 且}0\leqslant x_{ij}\leqslant C_{ij},\text{那么给}V_j标号(-V_i,\Delta_i)，\Delta_i=min\{x_{ij},\Delta_i\} \\
&\phantom{xxxx}\text{重复步骤}2,\text{直到}V_t\text{被标号或标号过程无法进行下去}\\
&\text{若}V_t\text{被标记},\text{则有一条增广链},\text{转调整过程}; \\
&\phantom{xxxx}\phantom{xxxx}取增广链中最小的 \Delta \\
&\phantom{xxxx}\phantom{xxxx}\text{令}x_{ij}=\begin{cases}x_{ij}+\Delta,\text{链}u\text{ 的前向弧}\\x_{ij}-\Delta,\text{链}u\text{ 的后向弧,2}.\text{去掉所有标号,回到第一步},\text{对可行流进行重新标号}.\\x_{ij},\text{弧}\notin\text{链}u&\end{cases}\\
&\text{若}V_t\text{未被标记},\text{而过程无法进行下去,这时可行流就是最大流}.
\end{aligned}$$

```python
maximum_flow (flowG, _s, _t, capacity=‘capacity’, flow_func=None, *kwargs)  
maximum_flow_value (flowG, _s, _t, capacity=‘capacity’, flow_func=None, *kwargs)
```
* flowG(NetworkX graph)：有向图，边必须带有容量属性 capacity（不能用 ‘weight’ ）。
* _s (node)：源点
* _t (node)：汇点
* capacity (string)：边的容量属性 capacity，缺省视为无限容量。
* flow_func(function)：调用算法的函数名，如：'edmonds_karp','shortest_augmenting_path','dinitz','preflow_push', 'boykov_kolmogorov'。缺省值'None'，选择'preflow_push'（HLPP 算法）

```python
# 调用其他的算法
from networkx.algorithms.flow import edmonds_karp
flowValue, flowDict = nx.maximum_flow(G1, 's', 't', flow_func=edmonds_karp) 
```

````python
# 例：
import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms.flow import edmonds_karp  # 导入 edmonds_karp 算法函数

# 1. 最大流问题 (Maximum Flow Problem，MFP)
# 创建有向图
G1 = nx.DiGraph()  # 创建一个有向图 DiGraph
G1.add_edge('s', 'a', capacity=6)  # 添加边的属性 "capacity"
G1.add_edge('s', 'c', capacity=8)
G1.add_edge('a', 'b', capacity=3)
G1.add_edge('a', 'd', capacity=3)
G1.add_edge('b', 't', capacity=10)
G1.add_edge('c', 'd', capacity=4)
G1.add_edge('c', 'f', capacity=4)
G1.add_edge('d', 'e', capacity=3)
G1.add_edge('d', 'g', capacity=6)
G1.add_edge('e', 'b', capacity=7)
G1.add_edge('e', 'j', capacity=4)
G1.add_edge('f', 'h', capacity=4)
G1.add_edge('g', 'e', capacity=7)
G1.add_edge('h', 'g', capacity=1)
G1.add_edge('h', 'i', capacity=3)
G1.add_edge('i', 'j', capacity=3)
G1.add_edge('j', 't', capacity=5)

# 求网络最大流1
# maxFlowValue = nx.maximum_flow_value(G1, 's', 't')  # 求网络最大流的值
# maxFlowValue, maxFlowDict = nx.maximum_flow(G1, 's', 't')  # 求网络最大流
# 求网络最大流2
# maxFlowDict={'s': {'a': 6, 'c': 8}, 'a': {'b': 3, 'd': 3}}
maxFlowValue, maxFlowDict = nx.maximum_flow(G1, 's', 't', flow_func=edmonds_karp)  # 求网络最大流

# 数据格式转换
edgeCapacity = nx.get_edge_attributes(G1, 'capacity')
edgeLabel = {}  # 边的标签
for i in edgeCapacity.keys():  # 整理边的标签，用于绘图显示：i=('a','b')
    edgeLabel[i] = f'c={edgeCapacity[i]:}'  # 边的容量{('a','b'):'c=xxx'}
edgeLists = []  # 最大流的边的 list
for i in maxFlowDict.keys():
    for j in maxFlowDict[i].keys():
        edgeLabel[(i, j)] += ',f=' + str(maxFlowDict[i][j])  # 取出每条边流量信息存入边显示值{('a','b'):'c=xxx,f=xxx'}
        if maxFlowDict[i][j] > 0:  # 网络最大流的边（流量>0）
            edgeLists.append((i, j))

# 输出显示
print("最大流值: ", maxFlowValue)
print("最大流的途径及流量: ", maxFlowDict)  # 输出最大流的途径和各路径上的流量
print("最大流的路径：", edgeLists)  # 输出最大流的途径

# 绘制有向网络图
fig, ax = plt.subplots(figsize=(8, 6))
pos = {'s': (1, 8), 'a': (6, 7.5), 'b': (9, 8), 'c': (1.5, 6), 'd': (4, 6), 'e': (8, 5.5),  # 指定顶点绘图位置
       'f': (2, 4), 'g': (5, 4), 'h': (1, 2), 'i': (5.5, 2.5), 'j': (9.5, 2), 't': (11, 6)}
edge_labels = nx.get_edge_attributes(G1, 'capacity')
ax.set_title("Maximum flow of petroleum network with NetworkX")  # 设置标题
nx.draw(G1, pos, with_labels=True, node_color='c', node_size=300, font_size=10)  # 绘制有向图，显示顶点标签
nx.draw_networkx_edges(G1, pos, edgelist=edgeLists, edge_color='m')  # 设置指定边的颜色、宽度
nx.draw_networkx_edge_labels(G1, pos, edgeLabel, font_color='navy')  # 显示边的标签：'capacity' + maxFlow
plt.axis('on')
plt.show()
````

### 最小费用流问题（MCP）
从发点到收点的费用最小（新增费用约束），且流量不超过每段路径的最大流量，如图(最大流量,所需费用)  


* 以费用为评价标准找出最短路径  
* 找出增广链  
    * 对于未走过的弧，$(C_{ij},x_{ij}=0)$，标号，找出最小的$\Delta$，记为该条路径通过的流量总数，计算该条路径新增的费用  
    $$F=\Delta_{min}*\sum_{i\in A}cost_i\\
    A——该路径中每段路径的集合$$  
    * 增删弧  
    $$\begin{align*}
0<x_{ij}<c_{ij}\text{时,该条弧上的流量可能增大也可能减小,} & \text{故在原来的基础上加一条反向弧,权值设为原弧的相反数;} \\
x_{ij}=c_{ij}\text{时,该条弧上流量已达到最大值,无法再增加,} & \text{流量只有减少的可能,所以删除原弧,添加一条反向弧,权值设为原弧的相反数}
\end{align*}$$
    

````python
# 例：
import matplotlib.pyplot as plt
import networkx as nx

g = nx.DiGraph()  # 创建一个有向图 DiGraph
g.add_edges_from([('s', 'v1', {'capacity': 13, 'weight': 7}),
                  ('s', 'v2', {'capacity': 9, 'weight': 9}),
                  ('v1', 'v3', {'capacity': 6, 'weight': 6}),
                  ('v1', 'v4', {'capacity': 5, 'weight': 5}),
                  ('v2', 'v1', {'capacity': 4, 'weight': 4}),
                  ('v2', 'v3', {'capacity': 5, 'weight': 2}),
                  ('v2', 'v5', {'capacity': 5, 'weight': 5}),
                  ('v3', 'v4', {'capacity': 5, 'weight': 2}),
                  ('v3', 'v5', {'capacity': 4, 'weight': 1}),
                  ('v3', 't', {'capacity': 4, 'weight': 4}),
                  ('v4', 't', {'capacity': 9, 'weight': 7}),
                  ('v5', 't', {'capacity': 9, 'weight': 5})])

# 求最小费用最大流
# flowDict (dict)：字典类型，最小费用最大流的流经路径及各路径的分配流量
flowDict = nx.max_flow_min_cost(g, "s", "t")
# 由流经路径及各路径的分配流量 flowDict 计算可行流的成本
min_val = nx.cost_of_flow(g, flowDict)
print("最小费用最大流" + str(flowDict))
print("最小费用" + str(min_val))

# 输出标签(最大流量,所需费用)
cap = nx.get_edge_attributes(g, "capacity")
wei = nx.get_edge_attributes(g, "weight")
labels = {}
for label in cap.keys():
    labels[label] = "(" + str(cap[label]) + "," + str(wei[label]) + ")"

# 输出最小费用最大流real=xxx
for i in flowDict.keys():
    for j in flowDict[i].keys():
        if flowDict[i][j] > 0:
            labels[(i, j)] += ",real=" + str(flowDict[i][j])
nx.draw(g, pos=nx.shell_layout(g), with_labels=True, node_color='c', node_size=300, font_size=10)  # 绘制有向图，显示顶点标签
nx.draw_networkx_edge_labels(g, pos=nx.shell_layout(g), edge_labels=labels,
                             font_color='navy')  # 显示边的标签：'capacity' + maxFlow
plt.show()
````
