解决问题：确定一个系统中哪些因素是主要因素，哪些是次要因素  
原理：根据序列曲线的形状相似程度判断联系是否紧密  
* 数据预处理：正向化、标准化（见TOPSIS）  
* 确定子序列、母序列  
    母序列/参考序列$x_0$：反应系统行为特征  
    子序列/比较序列$x_i$：影响系统行为的因素  
* 确定灰色关联系数  
    $$
    两级最小差：a=\min_s\min_t|x_0(t)-x_s(t)|\\
    两级最大差：a=\max_s\max_t|x_0(t)-x_s(t)|\\
    灰色关联系数：\gamma(x_0(k),x_i(k))=\frac{a+\rho b}{|x_0(k)-x_i(k)|+\rho b}--（\rho 通常为0.5）
    $$
    分别取平均值作为子序列的灰色关联度
