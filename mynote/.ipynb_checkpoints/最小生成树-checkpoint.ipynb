```python
# 
import matplotlib.pyplot as plt
import networkx as nx

G1 = nx.Graph()
G1.add_weighted_edges_from([(1, 2, 5), (1, 3, 6), (2, 4, 2), (2, 5, 12), (3, 4, 6),
                            (3, 6, 7), (4, 5, 8), (4, 7, 4), (5, 8, 1), (6, 7, 5),
                            (7, 8, 10)])
T = nx.minimum_spanning_tree(G1)  # 
print(T.edges(data=True))  # ,data=True 

mst1 = nx.tree.minimum_spanning_edges(G1, algorithm="kruskal")  # 
print(list(mst1))  # 
mst2 = nx.tree.minimum_spanning_edges(G1, algorithm="prim", data=False)  # data=False 
print(list(mst2))

# 
pos = {1: (1, 5), 2: (3, 1), 3: (3, 9), 4: (5, 5), 5: (7, 1), 6: (6, 9), 7: (8, 7), 8: (9, 4)}  # 
nx.draw(G1, pos=nx.shell_layout(G1), with_labels=True, node_color='c', alpha=0.8)  # 
labels = nx.get_edge_attributes(G1, 'weight')
nx.draw_networkx_edges(G1, pos=nx.shell_layout(G1), edgelist=T.edges, edge_color='b', width=4)  # 
nx.draw_networkx_edge_labels(G1, pos=nx.shell_layout(G1), edge_labels=labels, font_color='m')  # 
plt.show()
```

```python
# mathmodel18_v1.py
# Demo18 of mathematical modeling algorithm
# Demo of minimum spanning tree(MST) with NetworkX
# Copyright 2021 YouCans, XUPT
# Crated2021-07-10

import numpy as np
import matplotlib.pyplot as plt #  Matplotlib 
import networkx as nx  #  NetworkX 
from scipy.spatial.distance import pdist, squareform

# # 2. 
# nCities = input("Input number of cities (n>=10):")
# nCities = int(nCities)
nCities = 20
np.random.seed(1)
xPos = np.random.randint(0, 100, nCities)  #  [0,100) 
yPos = np.random.randint(0, 100, nCities)  #  Ncities 

posCity = []
G2 = nx.complete_graph(nCities)  # 
for node in G2.nodes():
    G2.add_node(node, pos=(xPos[node], yPos[node]))  #  pos
    posCity.append(G2.nodes[node]["pos"])  #  pos
'''
scipy.spatial.distance.pdict(ndarray)
ndarray

arr(ndarray)=[(1,1),(2,2),(3,3)]
:
for i in range(len(arr)):
    for j in range(len(arr)):
        res[i * len(arr) + j] = distance_ij(arr)
'''
'''
scipy.spatial.distance.squareform
MM_ij = distance_ij
'''
dist = squareform(pdist(np.array(posCity)))  # 
for u, v in G2.edges:
    G2.add_edge(u, v, weight=np.round(dist[u][v],decimals=1))  #  dist(u,v)

T = nx.minimum_spanning_tree(G2, algorithm='kruskal')  # 
print("\n:\n",G2._node)
print("\n:\n",sorted(T.edges(data=True)))  # data=True 
# mst = nx.tree.minimum_spanning_edges(G2, algorithm="kruskal") # 
# for edge in sorted(list(mst)):
#     print(edge)

fig, ax = plt.subplots(figsize=(8, 6))
node_pos = nx.get_node_attributes(G2, 'pos')  # 
nx.draw(G2,node_pos,with_labels=True,node_color='c',edge_color='silver',node_size=300,font_size=10,font_color='r',alpha=0.8)  # 
# nx.draw_networkx_labels(G2, node_pos, labels=node_pos, font_size=6, horizontalalignment='left', verticalalignment='top')  #  pos
# edge_col = ['red' if edge in T.edges() else 'silver' for edge in G2.edges()]  # 
# nx.draw_networkx_edges(G2, node_pos, edge_color=edge_col, width=2)  # 
nx.draw_networkx_edges(G2, node_pos, edgelist=T.edges, edge_color='r', width=2)  # 
edge_weight = nx.get_edge_attributes(T, 'weight')  # 
nx.draw_networkx_edge_labels(T, node_pos, edge_labels=edge_weight, font_size=8, font_color='m', verticalalignment='top')  # 
plt.axis('on')  # Remove the axis
plt.xlim(-5, 100)
plt.ylim(-5, 100)
plt.show()
```
